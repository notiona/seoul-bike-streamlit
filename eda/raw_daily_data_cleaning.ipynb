### Rename Korean columns to English

```python
rename_mapper = {
    "대여일자": "timestamp",
    "대여소번호": "rent_station_code",
    "대여소": "rent_station_name",
    "대여구분코드": "rent_type",
    "성별": "gender",
    "연령대": "age",
    " 이용건수": "rent_count",
    "운동량": "exercise_amount",
    "탄소량": "carbon_emission",
    "이동거리(M)": "rent_meter",
    "이용시간(분)": "rent_minute"
}
```

```python
dec_df.columns
```

```python
dec_df = dec_df.rename(columns=rename_mapper)
dec_df
```

### Split year, month, day from timestamp

```python
dec_df["timestamp"] = dec_df["timestamp"].astype("str")
dec_df["year"] = dec_df["timestamp"].str.slice(start=0, stop=4).astype("int32")
dec_df["month"] = dec_df["timestamp"].str.slice(start=5, stop=7).astype("int32")
dec_df["day"] = dec_df["timestamp"].str.slice(start=8).astype("int32")
dec_df
```

### Group by year, month and save

```python
for group, df in dec_df.groupby(["year", "month"]):
    print(group)
    year, month = group
    df = df.reset_index(drop=True)
    df.to_parquet(DAILY_DATA_OUTPUT / f"{year}_{str(month).zfill(2)}.parq")
```
